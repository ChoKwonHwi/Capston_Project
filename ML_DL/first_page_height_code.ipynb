```python
male_df['height'] = male_df.groupby('days')['height'].transform(lambda x: x.fillna(x.median()))
male_df['weight'] = male_df.groupby('days')['weight'].transform(lambda x: x.fillna(x.median()))
female_df['height'] = female_df.groupby('days')['height'].transform(lambda x: x.fillna(x.median()))
female_df['weight'] = female_df.groupby('days')['weight'].transform(lambda x: x.fillna(x.median()))
```

```python
# ì´ìƒì¹˜ ì œê±° í•¨ìˆ˜
def remove_outliers(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    df = df.loc[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df
```

```python
# ë‚¨ì„± ìƒí›„ ì¼ ìˆ˜ ë³„ë¡œ ê·¸ë£¹í™”
grouped_male = male_df.groupby('days')

# ê° ìƒí›„ ì¼ ìˆ˜ë³„ë¡œ ì´ìƒì¹˜ ì œê±°ëœ ë°ì´í„°í”„ë ˆì„ì„ ì €ì¥í•  ë¹ˆ ë°ì´í„°í”„ë ˆì„ ìƒì„±
cleaned_df_male = pd.DataFrame(columns=male_df.columns)

# ìƒí›„ ì¼ ìˆ˜(days)ë³„ë¡œ ì´ìƒì¹˜ ì œê±°ëœ ë°ì´í„°ë¥¼ cleaned_df_maleì— ì¶”ê°€
for name, group in grouped_male:
    group = remove_outliers(group, 'height')
    group = remove_outliers(group, 'weight')
    cleaned_df_male = pd.concat([cleaned_df_male, group])
```

```python
# ì—¬ì„± ìƒí›„ ì¼ ìˆ˜ ë³„ë¡œ ê·¸ë£¹í™”
grouped_female = female_df.groupby('days')

# ê° ìƒí›„ ì¼ ìˆ˜ë³„ë¡œ ì´ìƒì¹˜ ì œê±°ëœ ë°ì´í„°í”„ë ˆì„ì„ ì €ì¥í•  ë¹ˆ ë°ì´í„°í”„ë ˆì„ ìƒì„±
cleaned_df_female = pd.DataFrame(columns=female_df.columns)

# ìƒí›„ ì¼ ìˆ˜(days)ë³„ë¡œ ì´ìƒì¹˜ ì œê±°ëœ ë°ì´í„°ë¥¼ cleaned_df_femaleì— ì¶”ê°€
for name, group in grouped_female:
    group = remove_outliers(group, 'height')
    group = remove_outliers(group, 'weight')
    cleaned_df_female = pd.concat([cleaned_df_female, group])
```

### ë°ì´í„° ëª¨ë¸ë§

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# ë…ë¦½ ë³€ìˆ˜ì™€ ì¢…ì† ë³€ìˆ˜ ë¶„ë¦¬
X = baby_growth_df[['days','gender','weight']]
y = baby_growth_df['height']

# train/test ë°ì´í„° ë¶„ë¦¬, ì„±ë³„ ë¹„ì¤‘ ìœ ì§€
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=X['gender'], random_state=42)
```

### (1) LinearRegression height ì˜ˆì¸¡ ëª¨ë¸

```python
# linear regression ëª¨ë¸ í•™ìŠµ
lr = LinearRegression()
lr.fit(X_train, y_train)

# test ë°ì´í„°ë¡œ ì˜ˆì¸¡
y_pred = lr.predict(X_test)

# ëª¨ë¸ ì„±ëŠ¥ í‰ê°€
r2 = r2_score(y_test, y_pred)
print('R-squared:', r2)

# MSE í‰ê°€
mse = mean_squared_error(y_test, y_pred)
print("LinearRegression MSE:", mse)

# ì •í™•ë„ í‰ê°€
print("í›ˆë ¨ ì„¸íŠ¸ ì •í™•ë„: {:.3f}".format(lr.score(X_train, y_train)))
print("í…ŒìŠ¤íŠ¸ ì„¸íŠ¸ ì •í™•ë„: {:.3f}".format(lr.score(X_test, y_test)))
```

### (3) RandomForestRegressor - height ì˜ˆì¸¡ ëª¨ë¸

```python
# RandomForestRegressor ëª¨ë¸ ê°ì²´ ìƒì„±
rf = RandomForestRegressor(random_state=42)

# ëª¨ë¸ í•™ìŠµ
rf.fit(X_train, y_train)

# ëª¨ë¸ í‰ê°€ - ì˜ˆì¸¡ê°’ ìƒì„±
y_pred = rf.predict(X_test)

# ëª¨ë¸ í‰ê°€ - í‰ê·  ì œê³± ì˜¤ì°¨ ê³„ì‚°
mse = mean_squared_error(y_test, y_pred)
print("RandomForestRegressor Mean Squared Error:", mse)

# ì •í™•ë„ í‰ê°€
print("í›ˆë ¨ ì„¸íŠ¸ ì •í™•ë„: {:.3f}".format(rf.score(X_train, y_train)))
print("í…ŒìŠ¤íŠ¸ ì„¸íŠ¸ ì •í™•ë„: {:.3f}".format(rf.score(X_test, y_test)))
```

### (2) XGBRegressor - height ì˜ˆì¸¡ ëª¨ë¸

```python
# ëª¨ë¸ ê°ì²´ ìƒì„±
xgb_model = XGBRegressor(objective='reg:squarederror', tree_method='hist', random_state=42)

# ëª¨ë¸ í•™ìŠµ
xgb_model.fit(X_train, y_train)

# ê²€ì¦ìš© ë°ì´í„°ë¥¼ ì‚¬ìš©í•˜ì—¬ ì˜ˆì¸¡ ìˆ˜í–‰
y_pred = xgb_model.predict(X_test)

# ëª¨ë¸ ì„±ëŠ¥ í‰ê°€
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"xgboost MSE: {mse:.3f}")
print(f"xgboost MAE: {mae:.3f}")
print(f"xgboost R-squared: {r2:.3f}")

# ì •í™•ë„ í‰ê°€
print("í›ˆë ¨ ì„¸íŠ¸ ì •í™•ë„: {:.3f}".format(xgb_model.score(X_train, y_train)))
print("í…ŒìŠ¤íŠ¸ ì„¸íŠ¸ ì •í™•ë„: {:.3f}".format(xgb_model.score(X_test, y_test)))
```

### RandomForestRegressor hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV

# í•˜ì´í¼íŒŒë¼ë¯¸í„° í›„ë³´ ê°’ë“¤
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 5, 10]
}

# GridSearchCV ìˆ˜í–‰
rf = RandomForestRegressor(random_state=42)
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# ìµœì ì˜ í•˜ì´í¼íŒŒë¼ë¯¸í„°ì™€ ê·¸ ë•Œì˜ ì„±ëŠ¥ ì¶œë ¥
print("Best hyperparameters:", grid_search.best_params_)
print("Best CV score:", -grid_search.best_score_)

# ìµœì ì˜ í•˜ì´í¼íŒŒë¼ë¯¸í„°ë¡œ ëª¨ë¸ í•™ìŠµ ë° í‰ê°€
rf_best = RandomForestRegressor(n_estimators=grid_search.best_params_['n_estimators'],
                                 max_depth=grid_search.best_params_['max_depth'],
                                 min_samples_split=grid_search.best_params_['min_samples_split'],
                                 random_state=42)
rf_best.fit(X_train, y_train)
y_pred_rf = rf_best.predict(X_test)
print("Test MSE:", mean_squared_error(y_test, y_pred_rf))

# ì •í™•ë„ í‰ê°€
print("í›ˆë ¨ ì„¸íŠ¸ ì •í™•ë„: {:.3f}".format(rf_best.score(X_train, y_train)))
print("í…ŒìŠ¤íŠ¸ ì„¸íŠ¸ ì •í™•ë„: {:.3f}".format(rf_best.score(X_test, y_test)))
```
